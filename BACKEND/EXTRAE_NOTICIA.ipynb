```python
#Campo 8: Texto de la noticia
Texto_Noticia = requests.get(Enlace_Web)
soup = BeautifulSoup(Texto_Noticia.content,'html.parser')
#Texto_Noticia = soup.find_all(class_="epigraph paragraph")
Texto_Noticia = soup.find_all(class_="text_note_single md:border-0 md:p-0 dark:text-gray-200 md:dark:hover:text-white md:dark:hover:bg-transparent dark:border-gray-700")
Texto_Noticia = str(Texto_Noticia)

#Función para eliminar tags y su contenido en HTML
Eliminar_Tags = re.compile(r'<[^>]+>')
def remove_tags(text):
    text=text.replace("[", " " )
    text=text.replace("]", " " )
    return Eliminar_Tags.sub('', text)

Texto_Noticia=remove_tags(Texto_Noticia)
print("Texto Noticia: ",Texto_Noticia)
```

```python
#Campo 9: Acceso a la noticia. 
#Códigos de estados: 200(Success) - 400(Not-Found) - 
#1xx: Informational: It means the request was received and the process is continuing.
#2xx: Success: It means the action was successfully received, understood, and accepted.
#3xx: Redirection: It means further action must be taken in order to complete the request.
#4xx: Client Error: It means the request contains incorrect syntax or cannot be fulfilled.
#5xx: Server Error: It means the server failed to fulfill an apparently valid request.
Acceso_Noticia = ""
Texto_Noticia = requests.get(Enlace_Web)
if Texto_Noticia.status_code == 200:
    print("Código de estado: ",Texto_Noticia.status_code)
    Acceso_Noticia = "SI"
else:
    print("Página no encontrada. Código de error: ",Texto_Noticia.status_code)
    Acceso_Noticia = "NO"
print("Acceso a la noticia: ",Acceso_Noticia)
```
