```python
import getpass
import pymongo

from random import randint
from bson.objectid import ObjectId
```

```python
def read_all_document(coleccion):
    """Return the contents of all documents"""
    return coleccion.find()

def read_specific_document_id(coleccion, _id):
    """Return the contents of the document containing document_id"""
    return coleccion.find( {"_id": ObjectId(_id) })

def read_specific_document_fecha(coleccion, fecha_noticia):
    """Return the contents of the document containing document_id"""
    return coleccion.find( {"fecha_noticia": fecha_noticia })

def read_specific_document_actividad(coleccion, actividad_economica):
    """Return the contents of the document containing document_id"""
    return coleccion.find( {"actividad_economica": actividad_economica })

def read_specific_document_noticia(coleccion, noticia):
    """Return the contents of the document containing document_id"""
    return coleccion.find( {"noticia": noticia })

def read_specific_document_cobertura(coleccion, cobertura):
    """Return the contents of the document containing document_id"""
    return coleccion.find( {"cobertura": cobertura })
```
