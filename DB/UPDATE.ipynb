```python
def update_specific_document_fecha(coleccion, _id, fecha_noticia):
    """Return the contents of the document containing document_id"""
    myquery = { "_id": _id }
    newvalues = { "$set": { "fecha_noticia": fecha_noticia } }
    coleccion.update_one(myquery, newvalues)
    print("Updated document with id {}".format(_id))

def update_specific_document_actividad(coleccion, _id,  actividad_economica):
    """Return the contents of the document containing document_id"""
    myquery = { "_id": _id }
    newvalues = { "$set": { "actividad_economica": actividad_economica } }
    coleccion.update_one(myquery, newvalues)
    print("Updated document with id {}".format(_id))

def update_specific_document_noticia(coleccion, _id,  noticia):
    """Return the contents of the document containing document_id"""
    myquery = { "_id": _id }
    newvalues = { "$set": { "noticia": noticia } }
    coleccion.update_one(myquery, newvalues)
    print("Updated document with id {}".format(_id))

def update_specific_document_cobertura(coleccion, _id,  cobertura):
    """Return the contents of the document containing document_id"""
    myquery = { "_id": _id }
    newvalues = { "$set": { "cobertura": cobertura } }
    coleccion.update_one(myquery, newvalues)
    print("Updated document with id {}".format(_id))
```
